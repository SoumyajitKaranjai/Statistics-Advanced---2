Question 1: What is hypothesis testing in statistics?

Hypothesis testing in statistics is a formal procedure used to make inferences or draw conclusions about a population parameter by analyzing data from a sample.

The process starts with an assumption called the null hypothesis (H0)
, typically stating that there is no effect or no difference in the population. Researchers also define an alternative hypothesis (H1 or
Ha), which suggests the opposite—for instance, that an effect or difference does exist.

Key steps in hypothesis testing include:

a. Stating the hypotheses (null and alternative).

b. Choosing a significance level (α), often 0.05, which specifies the maximum probability of incorrectly rejecting the null hypothesis.

c. Selecting an appropriate statistical test (such as t-test, z-test, or chi-square test) based on data and hypothesis type.

d. Calculating a test statistic that quantifies the difference between observed data and what is expected under H0.

e. Computing a p-value, which measures the probability of observing data as extreme as what was collected, assuming the null hypothesis is true.

f. Making a decision: If the p-value is less than or equal to
α, reject the null hypothesis; otherwise, do not reject it.

Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?

**Null Hypothesis:-**
The null hypothesis (denoted as H0) is a default statistical assumption that states there is no effect, no difference, or no relationship between groups, variables, or conditions in a population—the observed data are purely due to chance. It serves as the baseline assumption to be tested, often formulated as “no change” or “no association.”

**Alternative Hypothesis:-**
The alternative hypothesis (denoted as Ha ) asserts that there is an effect, difference, or relationship—that some change or association exists in the population.

Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

**Signifance Level:-**

The significance level in hypothesis testing, denoted as α (alpha), is the probability of incorrectly rejecting the null hypothesis when it is actually true (a Type I error). It sets the threshold for statistical evidence required to conclude that an observed effect is real rather than due to chance.

Role in deciding the outcome:

*  After analyzing the data, a p-value is calculated, representing the probability of observing results at least as extreme as those found, assuming the null hypothesis is true.

*   If the p-value is less than or equal to α, you reject the null hypothesis and conclude the result is statistically significant.
*   If the p-value is greater than α, you fail to reject the null hypothesis.

Question 4: What are Type I and Type II errors? Give examples of each.

Type I and Type II errors are two kinds of incorrect conclusions that can occur in statistical hypothesis testing:

**Type I error** (false positive): This occurs when you reject the null hypothesis (H0) even though it is actually true. In other words, you conclude that there is an effect or difference when in reality there is none.

Example: Suppose a medical test is designed to detect a disease. The null hypothesis states the patient does not have the disease. If the test incorrectly indicates the patient does have the disease (when they really do not), this is a Type I error—a false alarm.

**Type II error** (false negative): This occurs when you fail to reject the null hypothesis (H0) even though it is actually false. That means you miss detecting an effect or difference that is truly there.

Example: Using the same medical test scenario, the null hypothesis says the patient does not have the disease. If the test fails to detect the disease in a patient who actually is sick, this is a Type II error—missing a real case.

Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.

**Z-test:-**

*  Population standard deviation is known.

*   Sample size is large (typically n ≥ 30).

*   Test statistic distribution is Standard normal distribution (Z-distribution).
*   Population is normally distributed.

**T-test:-**

*   Population standard deviation is unknown.
*   Sample size is small (typically n < 30).

*   Test statistic distribution is T-distribution.
*   Population is normally distributed.

Use a Z-test when:

a. The sample size is large (generally 30 or greater).

b. The population standard deviation (σ) is known.

c. You are comparing means when these conditions are met.

Use a T-test when:

a. The sample size is small (less than 30).

b. The population standard deviation is unknown (must estimate using sample data).

c. You are comparing means under uncertainty about σ or with small samples.

Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.










```python
import numpy as np
import matplotlib.pyplot as plt

n = 10       # number of trials
p = 0.5      # probability of success
size = 1000  # number of random samples to generate

# Generate binomial distributed data
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(n+2), align='left', edgecolor='black', alpha=0.7)
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title(f'Histogram of Binomial Distribution (n={n}, p={p})')
plt.xticks(range(n+1))
plt.show()
```

Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in
Python. Show the Python code and interpret the results.
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
50.3, 50.4, 50.0, 49.7, 50.5, 49.9]


```python
import numpy as np
from statsmodels.stats.weightstats import ztest

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesized population mean
pop_mean = 50

# Known population standard deviation (assumed)
pop_std = 0.5

# Sample size
n = len(sample_data)

# Sample mean
sample_mean = np.mean(sample_data)

# Calculate standard error
std_error = pop_std / np.sqrt(n)

# Calculate Z-statistic
z_stat = (sample_mean - pop_mean) / std_error

# Calculate p-value for two-tailed test
from scipy.stats import norm
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Significance level
alpha = 0.05

# Output results
print(f"Sample mean: {sample_mean:.4f}")
print(f"Z-statistic: {z_stat:.4f}")
print(f"P-value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference from the population mean.")
else:
    print("Fail to reject the null hypothesis: No significant difference from the population mean.")
```

Interpretation:

a. The Z-statistic quantifies how many standard errors the sample mean is away from the hypothesized population mean.

b. The P-value tells us the probability of observing a sample mean at least as extreme as ours, assuming the null hypothesis (the true mean equals the population mean) is true.

c. If the p-value is less than the chosen significance level (0.05 here), we reject the null hypothesis, meaning there is statistically significant evidence that the sample mean differs from the population mean.

d. Otherwise, we do not reject the null hypothesis, indicating insufficient evidence to say the mean differs.

This approach assumes the population standard deviation is known, which justifies the use of a Z-test over a T-test.

Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.





```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

mean = 50          # population mean
std_dev = 5        # population standard deviation
sample_size = 100  # number of data points

# Simulate data from a normal distribution
data = np.random.normal(mean, std_dev, sample_size)

# Calculate sample mean and standard error of the mean
sample_mean = np.mean(data)
sem = stats.sem(data)  # standard error of the mean

# Calculate 95% confidence interval for the mean
confidence_level = 0.95
ci = stats.t.interval(confidence_level, df=sample_size-1, loc=sample_mean, scale=sem)

print(f"Sample mean: {sample_mean:.3f}")
print(f"95% confidence interval for the mean: ({ci[0]:.3f}, {ci[1]:.3f})")

# Plot histogram of the data
plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='blue', linestyle='dashed', linewidth=2, label='Sample Mean')
plt.axvline(ci[0], color='red', linestyle='dashed', linewidth=2, label='95% CI Lower Bound')
plt.axvline(ci[1], color='red', linestyle='dashed', linewidth=2, label='95% CI Upper Bound')
plt.title('Histogram of Simulated Normal Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_z_scores(data):
    # Calculate mean and standard deviation of the dataset
    mean = np.mean(data)
    std_dev = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std_dev

    # Plot the histogram of Z-scores
    plt.hist(z_scores, bins=15, edgecolor='black', alpha=0.7)
    plt.title('Histogram of Z-scores (Standardized Data)')
    plt.xlabel('Z-score')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()

    return z_scores

# Example usage
sample_data = np.array([49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7,
                        50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6,
                        50.0, 49.7, 50.2, 49.5, 50.1, 50.3, 50.4, 50.5,
                        50.0, 50.7, 49.3, 49.8, 50.2, 50.9, 50.3, 50.4,
                        50.0, 49.7, 50.5, 49.9])

z_scores = calculate_and_plot_z_scores(sample_data)
print("Z-scores:\n", z_scores)
```

Explanation:

a. A Z-score indicates how many standard deviations a particular data point is from the mean of the dataset.

b. A Z-score of 0 means the data point is exactly at the mean.

c.A positive Z-score means the data point is above the mean.

d. A negative Z-score means the data point is below the mean.

e. For example, a Z-score of 2 means the data point is 2 standard deviations above the mean, and a Z-score of -1.5 means it is 1.5 standard deviations below the mean.

f. Z-scores standardize different datasets for comparison by putting them on the same scale with mean 0 and standard deviation 1.

This function will help visualize how the data points are spread relative to the mean in terms of standard deviations.
